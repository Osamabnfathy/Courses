> 1-Encapsolation:
  - Access Modifier:
    - Public
    - Protected  "_" 
    - Private    "__"

```python
class Car:
  
    def __init__(self, speed : int = 250, color : str = "Blue", model : str = "BMW"):
        self.speed = speed    # Public
        self._color = color   # Protected
        self.__model = model  # Private
    
    
    def get_model(self):
        print(f"Car model is {self.__model}")
    
    def set_model(self, new_model):
        self.__model = new_model
        
    def turn_on(self):
        print(f"{self.__model} is turning on")
      
    def brake(self):
        print("Brake")
      
    def turn_off(self):
        print(f"{self.__model} is turning off")
      


Guilia = Car(color="Black", model="Alfa Romeo Giulia")
print(Guilia.speed)      # Give defaults "250"
Guilia.turn_on()

# protected property did't show but if you type it, it will work, and you can change it
print(Guilia._color)        
Guilia._color = "Yellow"
print(Guilia._color)        

# print(Guilia.__model)     # Private property
# Guilia.__model = "Suberu" # Didn't work
Guilia.get_model()          # Get data from Private Preperty 
Guilia.set_model("Suberu")  # Set data to the Private Preperty
Guilia.get_model()          # Get data from Private Preperty
```

> 2-Inheritance:
  - We have Super class "Father" and subclass "child"
  - Subclass inherite all Methods and attributes of the Super Class
  - Relation between 2 Classes "Is-A"

```python
class Person:
    def __init__(self, name : str = "name", email : str = "email", age : int = "0", hieght : int = "0", grade : str = "grade"):
        self.__age = age
        self.__name = name
        self.__email = email
        self.__grade = grade
        self.__hieght = hieght

    def get_age(self):
        return self.__age
    
    def set_age(self, new_age):
        self.__age = new_age

    def get_name(self):
        return self.__name
    
    def set_name(self, new_name):
        self.__name = new_name
    
    def get_email(self):
        return self.__email
    
    def set_email(self, new_email):
        self.__email = new_email
    
    def get_grade(self):
        return self.__grade
    
    def set_grade(self, new_grade):
        self.__grade = new_grade
    
    def get_hieght(self):
        return self.__hieght
    
    def set_hieght(self, new_hieght):
        self.__hieght = new_hieght
        
    def Eat(self):
        print("I'm eating")
    
    

# inherite from Person class
# has "Had_melitiray_service" attribute more than the Person class
class Boy(Person):
    
# these func and attributes no one can see it
# the constructor here depend on the constructor of the Person class
    def __init__(self, name: str = "name", email: str = "email", age: int = "0",hieght: int = "0", grade: str = "grade",
                 Had_melitiray_service : bool = "True"):
        super().__init__(name, email, age, hieght, grade)
        self.__melitiray_service = Had_melitiray_service
        
    def get_status(self):
        return self.__melitiray_service
    
    def set_status(self, new_condition):
        self.__melitiray_service = new_condition

    def Eat(self):
        print("As a Boy")
        # Func from Person class
        super().Eat()        
    

# inherite from Person class
# has "favshow" attribute more than the Person class
class Girl(Person):
    def __init__(self, name: str = "name", email: str = "email", age: int = "0", hieght: int = "0", grade: str = "grade", 
                 favshow : str = 'ay7aga'):
        super().__init__(name, email, age, hieght, grade)
        self.__favshow = favshow
        
    def get_show(self):
        return self.__favshow
    
    def set_show(self, new_show):
        self.__favshow = new_show
        

Per = Person()
Per.Eat()
print("\n")

Os = Boy(name="Os", age=21, Had_melitiray_service=False)
Os.get_name()
Os.get_age()
Os.get_status()
Os.Eat()
print("\n")

Ay = Girl(favshow="RDR")
Ay.Eat()
```

> Composition:
  - Relation between 2 Objects "Has-A" 

```python
class Car():
    def __init__(self, model, color, doors, engine):
        self.__model = model
        self.__color = color
        self.__doors = doors
        # composition of the Engine in the Car Class
        self.__engine = engine

    # Getter and setter for model attribute
    def get_model(self):
        return self.__model
    
    def set_model(self, new_model):
        self.__model = new_model
        
    # Getter and setter for color attribute     
    def get_color(self):
        return self.__color
    
    def set_color(self, new_color):
        self.__color = new_color
    
    # Getter and setter for doors attribute     
    def get_doors(self):
        return self.__doors
    
    def set_doors(self, new_num):
        self.__doors = new_num        
    
    # Getter and setter for engine attribute 
    def get_engine(self):
        return self.__engine
    
    def set_engine(self, new_engine):
        self.__engine = new_engine



class Engine():
    def __init__(self, type : str, power : int):
        self.__type = type
        self.__power = power
        
    # Getter and setter for type attribute   
    def get_type(self):
        return self.__type
    
    def set_type(self, new_type):
        self.__type = new_type
        
    # Getter and setter for power attribute   
    def get_power(self):
        return self.__power
    
    def set_power(self, new_power):
        self.__power = new_power
        
        

# composition of the Engine class with Car Class 
engine = Engine(type="V8", power=400)
car = Car("Model S", "Black", 4, engine)

print(car.get_model())
print(car.get_color())
print(car.get_doors())
print(car.get_engine().get_type())
print(car.get_engine().get_power())
```

> 3-Abstraction:
  - Hide Implementation
  - There are 2 types of the Abstraction "Interface - Abstract class"
  - Abstract class and an Interface class are both used to define a set of methods that a subclass must implement.
    - Interface class :
      - define abstract methods without any implementation.
      - define method signatures but no implementation details
      - define contracts that a class must adhere to
    - Abstract class :
      - can include both method signatures and implementation detail
      - define a common base class for a set of subclasses.
      - have both abstract methods and concrete methods
        - Abstract methods have no implementation and must be implemented by the subclass
        - concrete methods have an implementation and can be inherited by the subclass

  - A class can implement multiple interfaces, but can only inherit from one abstract class.
  - Can't instantiate from the abstract class
  - Inherite func from it then  make a body for it
  - The Porpuse is not to create Objects just for Inherite from it

```python
# Interface:

from abc import ABC, abstractmethod

# Interface class
class Animal(ABC):
    @abstractmethod
    def drink(self):
        pass

    @abstractmethod
    def walk(self):
        pass


class Dog(Animal):
    def drink(self):
        print("Dog is Drinking")

    def walk(self):
        print("Dog is walking")



class Cat(Animal):
    def drink(self):
        print("Cat is Drinking")
    
    def walk(self):
        print("Cat is walking")
    
    
Dog1 = Dog()
Dog1.walk()
Cat1 = Cat()
Cat1.walk()
```

```python
# Abstract:

from abc import ABC, abstractmethod

# Abstract Class:
class Animal(ABC):
    @abstractmethod
    def make_sound(self):
        pass

    # concrete method
    def eat(self, food):
        print(f"Animal is eating {food}")


class Dog(Animal):
    def make_sound(self):
        print("Woof!")


class Cat(Animal):
    def make_sound(self):
        print("Meow!")
        


Animal_1 = Cat()
Animal_1.eat(food="milk")   # Concrete method
Animal_1.make_sound()       # Abstract method

Animal_2 = Dog()
Animal_2.eat(food="meat")   # Concrete method
Animal_2.make_sound()       # Abstract method
```

> 4-Polymorphism:
  - Polymorphism is the ability of objects of different classes to be treated as if they were of the same type.
  - polymorphism is often achieved through method overloading or method overriding.

```python
# polymorphism using method overriding:

class Animal:
    def __init__(self, name):
        self.name = name

    def make_sound(self):
        print("This animal doesn't make any sound.")

class Dog(Animal):
    def make_sound(self):
        print("Woof!")

class Cat(Animal):
    def make_sound(self):
        print("Meow!")

animals = [Dog("Fido"), Cat("Whiskers"), Animal("Something")]

for animal in animals:
    animal.make_sound()
```

```python
# polymorphism in Python using method overloading:

class Calculator:
    def add(self, x, y):
        return x + y

    def add(self, x, y, z):
        return x + y + z

c = Calculator()

# print(c.add(1, 2))   # Give Error
print(c.add(1, 2, 3))
```
